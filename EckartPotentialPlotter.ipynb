```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
plt.rcParams.update({'font.size': 27})
```

```python
def eckart(x, A, B):
    return((-A*np.exp(x)/(1+np.exp(x)))-(B*np.exp(x)/((1+np.exp(x))**2)))
```

```python
def eckart_plotter(A, B):
    xvalues = np.linspace(-5, 5, 1000)
    yvalues = np.zeros(len(xvalues))
    for i in range(len(xvalues)):
        yvalues[i] = eckart(xvalues[i], A, B)
    plt.figure(figsize=(9, 6))
    plt.plot(xvalues+0.313, yvalues)
    plt.xlabel('Reaction coordinate (arb. units)')
    plt.ylabel('Energy (arb. units)')
    plt.yticks(np.linspace(-3, 3, 7))
    plt.grid()
    plt.savefig('./Figures/EckartPlot.svg', bbox_inches='tight')
```

```python
eckart_plotter(3.1, -20)
```
