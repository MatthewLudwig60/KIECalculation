# Calculating moments of inertia

1) Find the centre of mass

2) Convert all atomic coordinates to centre of mass coordinates

3) Diagonalise the matrix below

$I = \begin{pmatrix}I_{xx}&I_{xy}&I_{xz}\\I_{yx}&I_{yy}&I_{yz}\\I_{zx}&I_{zy}&I_{zz}\end{pmatrix} = \begin{pmatrix}\sum_i m_i(y_i^2+z_i^2)&-\sum_i m_i(x_iy_i)&-\sum_i m_i(x_iz_i)\\-\sum_i m_i(y_ix_i)&\sum_i m_i(x_i^2+z_i^2)&-\sum_i m_i(y_iz_i)\\-\sum_i m_i(z_ix_i)&-\sum_i m_i(z_iy_i)&\sum_i m_i(x_i^2+y_i^2)\end{pmatrix}$

## Using pre-IRC data and conditions 1+2

```python
def calculateKIE(r_atoms_path, TS_atoms_path, r_vib_path, TS_vib_path):
    r_atoms = pd.read_csv(r_atoms_path)
    TS_atoms = pd.read_csv(TS_atoms_path)
    r_vib = pd.read_csv(r_vib_path)
    TS_vib = pd.read_csv(TS_vib_path)
    
    #Calculation of COM and moments of inertia 
    
    r1_COM = np.zeros(3)
    r1_COM[0] = (np.sum(r_atoms['X_R1']*r_atoms['AtomicMassesR1']))/(np.sum(r_atoms['AtomicMassesR1']))
    r1_COM[1] = (np.sum(r_atoms['Y_R1']*r_atoms['AtomicMassesR1']))/(np.sum(r_atoms['AtomicMassesR1']))
    r1_COM[2] = (np.sum(r_atoms['Z_R1']*r_atoms['AtomicMassesR1']))/(np.sum(r_atoms['AtomicMassesR1']))

    r_atoms['X_R1COM'] = r_atoms['X_R1'] - r1_COM[0]
    r_atoms['Y_R1COM'] = r_atoms['Y_R1'] - r1_COM[1]
    r_atoms['Z_R1COM'] = r_atoms['Z_R1'] - r1_COM[2]
    
    Inertia_tensorR1 = np.zeros([3,3])
    Inertia_tensorR1[0][0] = np.sum(r_atoms['AtomicMassesR1']*u*(((r_atoms['Y_R1COM']*A)**2)+((r_atoms['Z_R1COM']*A)**2)))
    Inertia_tensorR1[1][1] = np.sum(r_atoms['AtomicMassesR1']*u*(((r_atoms['X_R1COM']*A)**2)+((r_atoms['Z_R1COM']*A)**2)))
    Inertia_tensorR1[2][2] = np.sum(r_atoms['AtomicMassesR1']*u*(((r_atoms['X_R1COM']*A)**2)+((r_atoms['Y_R1COM']*A)**2)))

    Inertia_tensorR1[1][0] = -np.sum(r_atoms['AtomicMassesR1']*u*r_atoms['Y_R1COM']*A*r_atoms['X_R1COM']*A)
    Inertia_tensorR1[0][1] = Inertia_tensorR1[1][0]

    Inertia_tensorR1[2][0] = -np.sum(r_atoms['AtomicMassesR1']*u*r_atoms['Z_R1COM']*A*r_atoms['X_R1COM']*A)
    Inertia_tensorR1[0][2] = Inertia_tensorR1[2][0]

    Inertia_tensorR1[2][1] = -np.sum(r_atoms['AtomicMassesR1']*u*r_atoms['Y_R1COM']*A*r_atoms['Z_R1COM']*A)
    Inertia_tensorR1[1][2] = Inertia_tensorR1[2][1]

    MomentsR1 = np.linalg.eig(Inertia_tensorR1)[0]
    print(f'The centre of mass of the reactant in condition 1 is {r1_COM} and its moments of inertia are {MomentsR1}')
   
    r2_COM = np.zeros(3)
    r2_COM[0] = (np.sum(r_atoms['X_R2']*r_atoms['AtomicMassesR2']))/(np.sum(r_atoms['AtomicMassesR2']))
    r2_COM[1] = (np.sum(r_atoms['Y_R2']*r_atoms['AtomicMassesR2']))/(np.sum(r_atoms['AtomicMassesR2']))
    r2_COM[2] = (np.sum(r_atoms['Z_R2']*r_atoms['AtomicMassesR2']))/(np.sum(r_atoms['AtomicMassesR2']))

    r_atoms['X_R2COM'] = r_atoms['X_R2'] - r2_COM[0]
    r_atoms['Y_R2COM'] = r_atoms['Y_R2'] - r2_COM[1]
    r_atoms['Z_R2COM'] = r_atoms['Z_R2'] - r2_COM[2]
    
    Inertia_tensorR2 = np.zeros([3,3])
    Inertia_tensorR2[0][0] = np.sum(r_atoms['AtomicMassesR2']*u*(((r_atoms['Y_R2COM']*A)**2)+((r_atoms['Z_R2COM']*A)**2)))
    Inertia_tensorR2[1][1] = np.sum(r_atoms['AtomicMassesR2']*u*(((r_atoms['X_R2COM']*A)**2)+((r_atoms['Z_R2COM']*A)**2)))
    Inertia_tensorR2[2][2] = np.sum(r_atoms['AtomicMassesR2']*u*(((r_atoms['X_R2COM']*A)**2)+((r_atoms['Y_R2COM']*A)**2)))

    Inertia_tensorR2[1][0] = -np.sum(r_atoms['AtomicMassesR2']*u*r_atoms['Y_R2COM']*A*r_atoms['X_R2COM']*A)
    Inertia_tensorR2[0][1] = Inertia_tensorR2[1][0]

    Inertia_tensorR2[2][0] = -np.sum(r_atoms['AtomicMassesR2']*u*r_atoms['Z_R2COM']*A*r_atoms['X_R2COM']*A)
    Inertia_tensorR2[0][2] = Inertia_tensorR2[2][0]

    Inertia_tensorR2[2][1] = -np.sum(r_atoms['AtomicMassesR2']*u*r_atoms['Y_R2COM']*A*r_atoms['Z_R2COM']*A)
    Inertia_tensorR2[1][2] = Inertia_tensorR2[2][1]

    MomentsR2 = np.linalg.eig(Inertia_tensorR2)[0]
    print(f'The centre of mass of the reactant in condition 2 is {r2_COM} and its moments of inertia are {MomentsR2}')
    
    TS1_COM = np.zeros(3)
    TS1_COM[0] = (np.sum(TS_atoms['X_TS1']*TS_atoms['AtomicMassesTS1']))/(np.sum(TS_atoms['AtomicMassesTS1']))
    TS1_COM[1] = (np.sum(TS_atoms['Y_TS1']*TS_atoms['AtomicMassesTS1']))/(np.sum(TS_atoms['AtomicMassesTS1']))
    TS1_COM[2] = (np.sum(TS_atoms['Z_TS1']*TS_atoms['AtomicMassesTS1']))/(np.sum(TS_atoms['AtomicMassesTS1']))

    TS_atoms['X_TS1COM'] = TS_atoms['X_TS1'] - TS1_COM[0]
    TS_atoms['Y_TS1COM'] = TS_atoms['Y_TS1'] - TS1_COM[1]
    TS_atoms['Z_TS1COM'] = TS_atoms['Z_TS1'] - TS1_COM[2]
    
    Inertia_tensorTS1 = np.zeros([3,3])
    Inertia_tensorTS1[0][0] = np.sum(TS_atoms['AtomicMassesTS1']*u*(((TS_atoms['Y_TS1COM']*A)**2)+((TS_atoms['Z_TS1COM']*A)**2)))
    Inertia_tensorTS1[1][1] = np.sum(TS_atoms['AtomicMassesTS1']*u*(((TS_atoms['X_TS1COM']*A)**2)+((TS_atoms['Z_TS1COM']*A)**2)))
    Inertia_tensorTS1[2][2] = np.sum(TS_atoms['AtomicMassesTS1']*u*(((TS_atoms['X_TS1COM']*A)**2)+((TS_atoms['Y_TS1COM']*A)**2)))

    Inertia_tensorTS1[1][0] = -np.sum(TS_atoms['AtomicMassesTS1']*u*TS_atoms['Y_TS1COM']*A*TS_atoms['X_TS1COM']*A)
    Inertia_tensorTS1[0][1] = Inertia_tensorTS1[1][0]

    Inertia_tensorTS1[2][0] = -np.sum(TS_atoms['AtomicMassesTS1']*u*TS_atoms['Z_TS1COM']*A*TS_atoms['X_TS1COM']*A)
    Inertia_tensorTS1[0][2] = Inertia_tensorTS1[2][0]

    Inertia_tensorTS1[2][1] = -np.sum(TS_atoms['AtomicMassesTS1']*u*TS_atoms['Y_TS1COM']*A*TS_atoms['Z_TS1COM']*A)
    Inertia_tensorTS1[1][2] = Inertia_tensorTS1[2][1]

    MomentsTS1 = np.linalg.eig(Inertia_tensorTS1)[0]
    print(f'The centre of mass of the TS in condition 1 is {TS1_COM} and its moments of inertia are {MomentsTS1}')
    
    TS2_COM = np.zeros(3)
    TS2_COM[0] = (np.sum(TS_atoms['X_TS2']*TS_atoms['AtomicMassesTS2']))/(np.sum(TS_atoms['AtomicMassesTS2']))
    TS2_COM[1] = (np.sum(TS_atoms['Y_TS2']*TS_atoms['AtomicMassesTS2']))/(np.sum(TS_atoms['AtomicMassesTS2']))
    TS2_COM[2] = (np.sum(TS_atoms['Z_TS2']*TS_atoms['AtomicMassesTS2']))/(np.sum(TS_atoms['AtomicMassesTS2']))
    
    TS_atoms['X_TS2COM'] = TS_atoms['X_TS2'] - TS2_COM[0]
    TS_atoms['Y_TS2COM'] = TS_atoms['Y_TS2'] - TS2_COM[1]
    TS_atoms['Z_TS2COM'] = TS_atoms['Z_TS2'] - TS2_COM[2]
    
    Inertia_tensorTS2 = np.zeros([3,3])
    Inertia_tensorTS2[0][0] = np.sum(TS_atoms['AtomicMassesTS2']*u*(((TS_atoms['Y_TS2COM']*A)**2)+((TS_atoms['Z_TS2COM']*A)**2)))
    Inertia_tensorTS2[1][1] = np.sum(TS_atoms['AtomicMassesTS2']*u*(((TS_atoms['X_TS2COM']*A)**2)+((TS_atoms['Z_TS2COM']*A)**2)))
    Inertia_tensorTS2[2][2] = np.sum(TS_atoms['AtomicMassesTS2']*u*(((TS_atoms['X_TS2COM']*A)**2)+((TS_atoms['Y_TS2COM']*A)**2)))

    Inertia_tensorTS2[1][0] = -np.sum(TS_atoms['AtomicMassesTS2']*u*TS_atoms['Y_TS2COM']*A*TS_atoms['X_TS2COM']*A)
    Inertia_tensorTS2[0][1] = Inertia_tensorTS2[1][0]

    Inertia_tensorTS2[2][0] = -np.sum(TS_atoms['AtomicMassesTS2']*u*TS_atoms['Z_TS2COM']*A*TS_atoms['X_TS2COM']*A)
    Inertia_tensorTS2[0][2] = Inertia_tensorTS2[2][0]

    Inertia_tensorTS2[2][1] = -np.sum(TS_atoms['AtomicMassesTS2']*u*TS_atoms['Y_TS2COM']*A*TS_atoms['Z_TS2COM']*A)
    Inertia_tensorTS2[1][2] = Inertia_tensorTS2[2][1]
    
    MomentsTS2 = np.linalg.eig(Inertia_tensorTS2)[0]
    print(f'The centre of mass of the TS in condition 2 is {TS2_COM} and its moments of inertia are {MomentsTS2}')
    
    #Calculation of partition function ratios 
    
    r_vib['nuR1'] = r_vib['VibR1']*((h*c)/(kB*T))
    r_vib['nuR2'] = r_vib['VibR2']*((h*c)/(kB*T))
    mass_ratio1 = (np.sum(r_atoms['AtomicMassesR2'])/np.sum(r_atoms['AtomicMassesR1']))**(3/2)
    inertia_prod1 = ((MomentsR2[0]*MomentsR2[1]*MomentsR2[2])/(MomentsR1[0]*MomentsR1[1]*MomentsR1[2]))**0.5
    vib_prod1 = 1
    for i in range(len(r_vib['nuR1'])):
        vib_prod1 = vib_prod1 * ((1-np.exp(-r_vib['nuR1'][i]))/(1-np.exp(-r_vib['nuR2'][i])))
    dummy = 0
    for i in range(len(r_vib['nuR1'])):
        dummy += (r_vib['nuR1'][i] - r_vib['nuR2'][i])/2
    vib_sum1 = np.exp(dummy)
    term1 = mass_ratio1 * inertia_prod1 * vib_prod1 * vib_sum1

    TS_vib['nuTS1'] = TS_vib['VibTS1']*((h*c)/(kB*T))
    TS_vib['nuTS2'] = TS_vib['VibTS2']*((h*c)/(kB*T))
    mass_ratio2 = (np.sum(TS_atoms['AtomicMassesTS1'])/np.sum(TS_atoms['AtomicMassesTS2']))**(3/2)
    inertia_prod2 = ((MomentsTS1[0]*MomentsTS1[1]*MomentsTS1[2])/(MomentsTS2[0]*MomentsTS2[1]*MomentsTS2[2]))**0.5
    vib_prod2 = 1
    for i in range(1, len(TS_vib['nuTS1'])):
        vib_prod2 = vib_prod2 * ((1-np.exp(-TS_vib['nuTS2'][i]))/(1-np.exp(-TS_vib['nuTS1'][i])))
    dummy = 0
    for i in range(1, len(TS_vib['nuTS1'])):
        dummy += (TS_vib['nuTS2'][i] - TS_vib['nuTS1'][i])/2
    vib_sum2 = np.exp(dummy)
    imag_ratio = TS_vib['nuTS1'][0]/TS_vib['nuTS2'][0]
    term2 = mass_ratio2 * inertia_prod2 * vib_prod2 * vib_sum2 * imag_ratio
    KIE = term1*term2
    
    print(f'KIE = {KIE}')
    return(KIE)
```

# Tunnelling Corrections

## Wigner
## $\kappa = 1 + \frac{(u^†)^2}{24}$

```python
def Wigner(u_A, u_B):
    kappa_A = 1 + (u_A**2)/24
    kappa_B = 1 + (u_B**2)/24
    KIE = kappa_A / kappa_B
    return(KIE)
```

## Bell
## $\alpha = \frac{E_b}{k_BT}$
## $\beta' = \frac{2\pi E_b}{h\nu^†}$
## If $\alpha > \beta': \kappa\approx\frac{\alpha e^{\alpha-\beta'}}{\alpha-\beta'}$
## If $\alpha\approx\beta': \kappa=\ln(1+e^{\beta'})$
## If $\alpha << \beta'$ and $\mu = \frac{h\nu^†}{k_BT} < 2\pi: \kappa = \frac{\mu/2}{\sin(\mu/2)}$ 

```python
def Bell(Eb_A, Eb_B, nu_A, nu_B): #input frequencies are dimensionless 
    alpha_A = Eb_A/(kB*T)
    beta_A = (2*np.pi*Eb_A)/(h*nu_A)
    alpha_B = Eb_B/(kB*T)
    beta_B = (2*np.pi*Eb_B)/(h*nu_B)
    
    kappa_A = 0
    kappa_B = 0
    
    if beta_A/alpha_A > 0.9 and beta_A/alpha_A < 1.1:
        kappa_A = np.log(1+np.exp(beta_A))
    else:
        if alpha_A > beta_A:
            kappa_A = (alpha_A*np.exp(alpha_A - beta_A))/(alpha_A - beta_A)
        else:
            mu_A = (h*nu_A)/(kB*T)
            if mu_A < 2*np.pi:
                kappa_A = (mu_A/2)/(np.sin(mu_A/2))
    
        
    if beta_B/alpha_B > 0.9 and beta_B/alpha_B < 1.1:
        kappa_B = np.log(1+np.exp(beta_B))
    else:
        if alpha_B > beta_B:
            kappa_B = (alpha_B*np.exp(alpha_B - beta_B))/(alpha_B - beta_B)
        else:
            mu_B = (h*nu_B)/(kB*T)
            if mu_B < 2*np.pi:
                kappa_B = (mu_B/2)/(np.sin(mu_B/2))
    if kappa_A == 0 or kappa_B == 0:
        KIE = 'Bell correction cannot be found'
    else:
        KIE = kappa_A/kappa_B
    return(KIE)  
```

## Central Differences Method based on Wigner 

## $\kappa = 1 - \frac{1}{24}(\frac{h}{k_BT})^2V_2 - \frac{h^2}{96\mu_rk_BT}\frac{V_4}{V_2}$
## $V_i = \frac{d^iV_{eff}(q_r)}{dq^i}\Bigg|_{q_r(TS)}$

```python
def iterate(q, V):
    V_prime = []
    q_prime = []
    for i in range(1, (len(q)-1)):
        V_prime.append((V[i+1] - V[i-1])/(q[i+1] - q[i-1]))
        q_prime.append(q[i])
    return(q_prime, V_prime)
```

```python
def calc_derivative(q, V, derivative):
    counter = 0
    while counter < derivative:
        q, V = iterate(q, V)
        counter += 1
    return(V[q.index(0)])
```

```python
def central_diff(q_A, q_B, V_A, V_B, mu_A, mu_B):
    V2_A = calc_derivative(q_A, V_A, 2)*hartree/(bohr**2)
    V2_B = calc_derivative(q_B, V_B, 2)*hartree/(bohr**2)
    V4_A = calc_derivative(q_A, V_A, 4)*hartree/(bohr**4)
    V4_B = calc_derivative(q_B, V_B, 4)*hartree/(bohr**4)
    
    kappa_A = 1-(1/24)*((h/(kB*T))**2)*V2_A - ((h**2)/(96*mu_A*kB*T))*(V4_A/V2_A)
    kappa_B = 1-(1/24)*((h/(kB*T))**2)*V2_B - ((h**2)/(96*mu_B*kB*T))*(V4_B/V2_B)
    
    print(kappa_A)
    print(kappa_B)
    
    KIE = kappa_A/kappa_B
    return(KIE)
```

## Eckart

```python
def eckart(V1_A, V1_B, V2_A, V2_B, nu_A, nu_B): #u is dimensionless imaginary barrier frequency, input barrier energies given in joules
    Data_X = [-.9324695,-.6612094,-.2386192,.2386192,.6612094,.9324695]
    Data_W = [.1713245,.3607616,.4679139,.4679139,.3607616,.1713245]
    
    alpha1_A = (2*np.pi*V1_A)/(h*nu_A)
    alpha1_B = (2*np.pi*V1_B)/(h*nu_B)
    alpha2_A = (2*np.pi*V2_A)/(h*nu_A)
    alpha2_B = (2*np.pi*V2_B)/(h*nu_B)
    
    u_A = nu_A*h/(kB*T)
    u_B = nu_B*h/(kB*T)
    
    c_A = (np.pi*u_A/8)*(((alpha1_A**-0.5)+(alpha2_A**-0.5))**2)
    c_B = (np.pi*u_B/8)*(((alpha1_B**-0.5)+(alpha2_B**-0.5))**2)
    
    v1_A = V1_A/(kB*T)
    v1_B = V1_B/(kB*T)
    v2_A = V2_A/(kB*T)
    v2_B = V2_B/(kB*T)
    
    d_A = 4*alpha1_A*alpha2_A-(np.pi)**2
    if d_A < 0:
        D_A = np.cos((-d_A)**0.5)
    else:
        D_A = np.cosh((d_A)**0.5)
    
    d_B = 4*alpha1_B*alpha2_B-(np.pi)**2
    if d_B < 0:
        D_B = np.cos((-d_B)**0.5)
    else:
        D_B = np.cosh((d_B)**0.5)
    
    if v2_A >= v1_A:
        EZ_A = -v1_A
    else:
        EZ_A = -v2_A
    
    if v2_B >= v1_B:
        EZ_B = -v1_B
    else:
        EZ_B = -v2_B
        
    EB_A = min((c_A*((np.log(2*(1+D_A)/0.014)/(2*np.pi))**2)-0.5*(v1_A + v2_A)), 3.2) 
    EB_B = min((c_A*((np.log(2*(1+D_B)/0.014)/(2*np.pi))**2)-0.5*(v1_B + v2_B)), 3.2)
    
    EM_A = 0.5*(EB_A - EZ_A)
    EM_B = 0.5*(EB_B - EZ_B)
    
    EP_A = 0.5*(EB_A + EZ_A)
    EP_B = 0.5*(EB_B + EZ_B)
    
    G_A = 0
    for i in range(len(Data_X)):
        E = EM_A*Data_X[i] + EP_A
        A1 = np.pi*(((E+v1_A)/c_A)**0.5)
        A2 = np.pi*(((E+v2_A)/c_A)**0.5)
        FP = np.cosh(A1+A2)
        FM = np.cosh(A1-A2)
        G_A += Data_W[i]*np.exp(-E)*(FP-FM)/(FP+D_A)
    G_A = EM_A*G_A+np.exp(-EB_A)
    
    G_B = 0
    for i in range(len(Data_X)):
        E = EM_B*Data_X[i] + EP_B
        A1 = np.pi*(((E+v1_B)/c_B)**0.5)
        A2 = np.pi*(((E+v2_B)/c_B)**0.5)
        FP = np.cosh(A1+A2)
        FM = np.cosh(A1-A2)
        G_B += Data_W[i]*np.exp(-E)*(FP-FM)/(FP+D_B)
    G_B = EM_B*G_B+np.exp(-EB_B)
    
    KIE = G_A/G_B
    return(KIE)
```

# Bigeleisen-Meyer Equation

## $\frac{k_{hyd}}{k_{deu}} = \frac{\kappa_{hyd}(T)}{\kappa_{deu}(T)}(\frac{k_{hyd}}{k_{deu}})_{SYM}(\frac{k_{hyd}}{k_{deu}})_{TS}(\frac{k_{hyd}}{k_{deu}})_{VPR}(\frac{k_{hyd}}{k_{deu}})_{EXC}(\frac{k_{hyd}}{k_{deu}})_{ZPVE}$
## $(\frac{k_{hyd}}{k_{deu}})_{SYM} = \frac{\sigma_{deu}^†}{\sigma_{hyd}^†}\frac{\sigma_{hyd}}{\sigma_{deu}}$
## $(\frac{k_{hyd}}{k_{deu}})_{TS} = \frac{\nu_{hyd}^†}{\nu_{deu}^†}$
## $(\frac{k_{hyd}}{k_{deu}})_{VPR} = \frac{\prod_j^{3N-6}u_{deu}/u_{hyd}}{\prod_j^{3N-7}u_{deu}^†/u_{hyd}^†}$
## $(\frac{k_{hyd}}{k_{deu}})_{EXC} = \frac{\prod_j^{3N-6}(1-e^{-u_{hyd}})/(1-e^{-u_{deu}})}{\prod_j^{3N-7}(1-e^{-u_{hyd}^†})/(1-e^{-u_{deu}^†})}$
## $(\frac{k_{hyd}}{k_{deu}})_{ZPVE} = e^{\frac{-1}{k_BT}((E_{deu,0} - E_{hyd,0})-(E_{deu,0}^† - E_{hyd,0}^†))}$ 
## $E_0 = \sum_j^{3N-6}\frac{h\nu_j}{2} = \sum_j^{3N-6}\frac{k_BTu_j}{2}$
## $E_0^† = \sum_j^{3N-7}\frac{h\nu_j^†}{2} = \sum_j^{3N-7}\frac{k_BTu_j^†}{2}$

```python
def BM(r_atoms_path, TS_atoms_path, r_vib_path, TS_vib_path, conditionA, conditionB, sym_num): #symmetry numbers given as array [condition A reactant, condition B reactant, condition A TS, condition B TS]

    r_atoms = pd.read_csv(r_atoms_path)
    TS_atoms = pd.read_csv(TS_atoms_path)
    r_vib = pd.read_csv(r_vib_path)
    TS_vib = pd.read_csv(TS_vib_path)
    
    r_vib[f'nuR{conditionA}'] = r_vib[f'VibR{conditionA}']*((h*c)/(kB*T))
    r_vib[f'nuR{conditionB}'] = r_vib[f'VibR{conditionB}']*((h*c)/(kB*T))
    TS_vib[f'nuTS{conditionA}'] = TS_vib[f'VibTS{conditionA}']*((h*c)/(kB*T))
    TS_vib[f'nuTS{conditionB}'] = TS_vib[f'VibTS{conditionB}']*((h*c)/(kB*T))
    
    #SYM
    SYM = (sym_num[3]/sym_num[2])*(sym_num[0]/sym_num[1])
    
    #TS
    TS = TS_vib[f'nuTS{conditionA}'][0]/TS_vib[f'nuTS{conditionB}'][0]
    
    #VPR
    VPR_R = 1
    for i in range(len(r_vib[f'nuR{conditionA}'])):
        VPR_R = VPR_R * (r_vib[f'nuR{conditionB}'][i]/r_vib[f'nuR{conditionA}'][i])
    VPR_TS = 1
    for i in range(1, len(TS_vib[f'nuTS{conditionA}'])):
        VPR_TS = VPR_TS * (TS_vib[f'nuTS{conditionB}'][i]/TS_vib[f'nuTS{conditionA}'][i])
    VPR = VPR_R/VPR_TS
    
    #EXC
    EXC_R = 1
    for i in range(len(r_vib[f'nuR{conditionA}'])):
        EXC_R = EXC_R * ((1-np.exp(-r_vib[f'nuR{conditionA}'][i]))/(1-np.exp(-r_vib[f'nuR{conditionB}'][i])))
    EXC_TS = 1
    for i in range(1, len(TS_vib[f'nuTS{conditionA}'])):
        EXC_TS = EXC_TS * ((1-np.exp(-TS_vib[f'nuTS{conditionA}'][i]))/(1-np.exp(-TS_vib[f'nuTS{conditionB}'][i])))
    EXC = EXC_R/EXC_TS
    
    #ZPVE
    E_RA = 0
    E_RB = 0
    for i in range(len(r_vib[f'nuR{conditionA}'])):
        E_RA += (kB*T/2)*r_vib[f'nuR{conditionA}'][i]
        E_RB += (kB*T/2)*r_vib[f'nuR{conditionB}'][i]
    E_TSA = 0
    E_TSB = 0
    for i in range(1, len(TS_vib[f'nuTS{conditionA}'])):
        E_TSA += (kB*T/2)*TS_vib[f'nuTS{conditionA}'][i]
        E_TSB += (kB*T/2)*TS_vib[f'nuTS{conditionB}'][i]
    ZPVE = np.exp((-1/(kB*T))*((E_RB - E_RA)-(E_TSB - E_TSA)))
    
    KIE = SYM*TS*VPR*EXC*ZPVE
    print(f'SYM = {SYM}')
    print(f'TS = {TS}')
    print(f'VPR = {VPR}')
    print(f'EXC = {EXC}')
    print(f'ZPVE = {ZPVE}')
    print(f'KIE = {KIE}')
    return(KIE)
```
