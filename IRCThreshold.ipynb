```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def BondLength(atomic_positions, IRC, donorC, acceptorC, targetH, num_points):
    data = pd.read_csv(atomic_positions)
    IRC_data = pd.read_csv(IRC)
    data_array = data.to_numpy()
    data_array = np.delete(data_array, [0,1], 1)
    
    donorBL = np.zeros(num_points)
    acceptorBL = np.zeros(num_points)
    
    for i in range(0, num_points*3, 3):
        donorBL[int(i/3)] = ((data_array[donorC - 1][i] - data_array[targetH - 1][i])**2 + \
        (data_array[donorC - 1][i+1] - data_array[targetH - 1][i+1])**2 + (data_array[donorC - 1][i+2]\
        - data_array[targetH - 1][i+2])**2)**0.5
        acceptorBL[int(i/3)] = ((data_array[acceptorC - 1][i] - data_array[targetH - 1][i])**2 + \
        (data_array[acceptorC - 1][i+1] - data_array[targetH - 1][i+1])**2 + (data_array[acceptorC - 1][i+2]\
        - data_array[targetH - 1][i+2])**2)**0.5
    IRC_data['Donor Bond Length'] = donorBL.tolist()
    IRC_data['Acceptor Bond Length'] = acceptorBL.tolist()
    return(IRC_data)
```

## C3

```python
a = BondLength('./OutputData/Morpholin2one/Reaction1/IRC/C3AtomicPositions.csv',\
               './OutputData/Morpholin2one/Reaction1/IRC/C3IRCEnergies.csv', 13, 4, 25, 41)
#atoms labelled as in Gaussian (i.e. indexed from 1 to 26)
```

```python
plt.figure(figsize=(12,6))
plt.rcParams.update({'font.size': 27})
plt.grid()
plt.plot(a['q1'], a['Donor Bond Length'], marker='o')
plt.plot(a['q1'], a['Acceptor Bond Length'], marker='o')
plt.axvline(a['q1'][14])
plt.axvline(a['q1'][31])
plt.legend(['Donor', 'Acceptor'])
```
